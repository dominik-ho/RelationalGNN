```python
import torch
from torch_geometric.data import Data
```

```python
edge_index = torch.tensor([[0, 1],
                           [1, 0],
                           [1, 2],
                           [2, 1]], dtype=torch.long)
edge_index = edge_index.t().contiguous()
edge_index
```

```python
row, col = edge_index
```

```python
row
```

```python
col
```

```python
from torch_geometric.utils import degree
```

```python
deg = degree(row, 3, dtype=torch.long)
print(deg)
deg_inv_sqrt = deg.pow(-0.5)
deg_inv_sqrt[deg_inv_sqrt == float('inf')] = 0
deg_inv_sqrt
```

```python
norm = deg_inv_sqrt[row] * deg_inv_sqrt[col]
norm.shape
```

```python
norm.view(-1, 1)
```

```python
x_j = torch.arange(0, 3*4, dtype=torch.float).view(4, 3)
x_j
```

```python
norm.view(-1, 1) * x_j
```
