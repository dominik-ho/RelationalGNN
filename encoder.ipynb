```python
import pymimir
from pymimir.pymimir.advanced.formalism import GroundAction

from graph_encoder import GraphEncoder
```

```python
d = pymimir.Domain("/home/dominik/PycharmProjects/RelationalGNN/data/domain.pddl")
p = pymimir.Problem(d, "/home/dominik/PycharmProjects/RelationalGNN/data/test_clear_probBLOCKS-3-0.pddl", "grounded")
```

```python
s = p.get_initial_state()
s
```

```python
# help(pymimir.GroundAction)
a1 = pymimir.GroundAction.new(d.get_action("pickup"), [p.get_object("a")], p)
a1
```

```python
s1 = a1.apply(s)
s1
```

```python
s
```

```python
a2 = pymimir.GroundAction.new(d.get_action("stack"), [p.get_object("a"), p.get_object("b")], p)
a2
```

```python
s2 = a2.apply(s1)
s2
```

```python
s = s2
```

```python
encoder = GraphEncoder(d, "obj")
```

```python
g = encoder.encode(s, p.get_objects())
g
```
